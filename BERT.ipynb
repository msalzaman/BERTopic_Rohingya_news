# BERTopic Modeling for Rohingya News
This project investigates Rohingya news framing by analyzing news published in 10 newspapers purposively selected from 10 relevant countries. The country selection relies on two categories: host countries and politically important countries. The analysis comparatively analyze Rohingya framing in the selected countries over the years, utilizing dynamic topic modeling (DTM) by BERT.

### 0. All countries
The corpus includes news items from the 10 selected countries. The analysis will focus on the following aspects:
* topics - 5 top topics
* topics over the years - 5 top topics
* keywords - 5 top keywords
* heatmap - 5 topics

```python
import pandas as pd
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

all_data = pd.read_excel('/Users/sayeed/Desktop/Datasets/cleaned_datasets/all_news.xlsx')

timestamps_all = all_data.date.to_list()
corpus_all = all_data.news.to_list()

vectorizer_model_all = CountVectorizer(stop_words="english")
topic_model_all = BERTopic(vectorizer_model=vectorizer_model_all, n_gram_range=(1, 2), min_topic_size=10, verbose=True)
topics, probs = topic_model_all.fit_transform(corpus_all)

%matplotlib inline
```

```python
topic_labels_all = {0: "repatriation", 1: "politics", 2: "resettlement", 3: "departure", 4: "relief"}
topic_model_all.set_topic_labels(topic_labels_all)
```

```python
top_topics_all = topic_model_all.get_topics(5)
# print(top_topics_all)
df_top_topics_all = pd.DataFrame(top_topics_all)
excel_file_path = '/Users/sayeed/Desktop/Datasets/analysis/all_topics.xlsx'
df_top_topics_all.to_excel(excel_file_path, index=False)
```

```python
topics_over_time_all = topic_model_all.topics_over_time(corpus_all, timestamps_all, nr_bins=5)
topic_model_all.visualize_topics_over_time(topics_over_time_all, top_n_topics=5, title='All countries', custom_labels=True)
```

```python
topic_model_all.visualize_barchart(top_n_topics=5, title='All countries', custom_labels=True)
```

```python
topic_model_all.visualize_heatmap(top_n_topics=5, title='All countries', custom_labels=True)
```
